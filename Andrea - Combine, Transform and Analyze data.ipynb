```python
def cal_percentages(series):
    state_dict = {}
    for state_Data in bd_series:
        len_state_users = len(state_Data[1])
        d = {"Platinum":0, "Gold":0, "Silver":0, "Bronze":0}
        for item in state_Data[1]:
            d[item] += 1
        for k,v in d.items():
            val = (v/len_state_users)*100
            d[k] = round(val,2)
        state_abbr = states_dict[state_Data[0]]
        state_dict[state_abbr] = d
    return state_dict
```

```python
full_data['fat'] = (full_data['WEIGHT']*0.453592) / ((full_data['HEIGHT']*0.0254)**2) 
full_data['fat'] = full_data['fat'].astype(int)
full_data['PURCHASED'] = full_data['PURCHASED'].apply(lambda x: x.upper())
full_data['s'] = full_data[full_data.PURCHASED == "SILVER"]["SILVER"]
full_data['g'] = full_data[full_data.PURCHASED == "GOLD"]["GOLD"]
full_data['p'] = full_data[full_data.PURCHASED == "PLATINUM"]["PLATINUM"]
full_data['b'] = full_data[full_data.PURCHASED == "BRONZE"]["BRONZE"]
full_data['PRE_CONDITIONS'] = full_data.fillna("Nothing")['PRE_CONDITIONS']
full_data['paid'] = ((full_data.fillna(0)['s'] + full_data.fillna(0)['g'] + full_data.fillna(0)['p'] + full_data.fillna(0)['b']) / 10).apply(lambda a: round(a, 0))
```

```python
full_data
```

```python
data_for_features_df = full_data[["TOBACCO","WEIGHT","ANNUAL_INCOME","PEOPLE_COVERED","OPTIONAL_INSURED","PURCHASED","paid","BRONZE","SILVER","GOLD","PLATINUM"]]
```

```python
data_for_features_df
```

```python
data_for_features_df["ANNUAL_INCOME"] = (data_for_features_df["ANNUAL_INCOME"] - data_for_features_df["ANNUAL_INCOME"].mean())/(data_for_features_df["ANNUAL_INCOME"].max() - data_for_features_df["ANNUAL_INCOME"].min())
```

```python
data_for_features_df["OPTIONAL_INSURED"] = (data_for_features_df["OPTIONAL_INSURED"] - data_for_features_df["OPTIONAL_INSURED"].mean())/(data_for_features_df["OPTIONAL_INSURED"].max() - data_for_features_df["OPTIONAL_INSURED"].min())
```

```python
data_for_features_df["WEIGHT"] = (data_for_features_df["WEIGHT"] - data_for_features_df["WEIGHT"].mean())/(data_for_features_df["WEIGHT"].max() - data_for_features_df["WEIGHT"].min())
```

```python
data_for_features_df
```

```python
data_for_features_df["TOBACCO"] = data_for_features_df["TOBACCO"].astype('category')
```

```python
data_for_features_df["PURCHASED"] = data_for_features_df["PURCHASED"].astype('category')
```
